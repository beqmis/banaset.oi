Banana Quality Dataset Analysis - December 2023

The Banana Quality Dataset provides detailed insights into the physical
and chemical attributes of bananas, alongside environmental factors influencing
their growth. This analysis is based on anonymized data collected from various regions,
varieties, and farming conditions.

Purpose: To analyze the factors influencing banana quality, including physical characteristics,
ripeness, and environmental conditions. This analysis aims to identify key drivers of quality and
provide actionable insights for optimizing banana production and improving agricultural outcomes.

Data Sources:
	•	Banana quality metrics (e.g., weight, length, firmness)
	•	Environmental conditions (e.g., soil nitrogen, rainfall, altitude)
	•	Data collected from regions like Colombia, Guatemala, and Ecuador

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
orig_data = pd.read_csv('/Users/akovdemidenko/Desktop/PyCharm Projects/bananchiki/.venv/banana_quality_dataset 2.csv')
orig_data
```

```python
orig_data.describe()
```

```python
orig_data.info()
```

```python
describe_data = orig_data[['quality_score','length_cm','weight_g','soil_nitrogen_ppm']]
mean_data = describe_data.mean()
std_data = describe_data.std()
median_data = describe_data.median()
skewness_data = describe_data.skew()

summary_data = pd.DataFrame({
    'Mean': mean_data,
    'Standard Deviation': std_data,
    'Median': median_data,
    'Skewness': skewness_data
})
summary_data
```

All four parameters show minimal skewness, indicating symmetrical distributions.

Quality scores have low variability, while length, weight, and soil nitrogen show moderate to high variability.

The dataset has no null values across all columns, ensuring its completeness. This makes it ready for analysis or modeling without requiring preprocessing to handle null entries.

```python
nan_count = orig_data.isnull().sum()
print(nan_count)
```

It illustrates that data is already clear and there is no NaN values.

The time-series plot highlights trends and variability in quality_score over harvest_date, providing key insights into temporal patterns.

```python
orig_data['harvest_date'] = pd.to_datetime(orig_data['harvest_date'])
sorted_data = orig_data.sort_values('harvest_date')
plt.figure(figsize=(10, 6))
sns.lineplot(data=sorted_data, x="harvest_date", y="quality_score", marker="o", color="purple")
plt.title("Time-Series Plot")
plt.xlabel("Harvest date")
plt.ylabel("Quality score")
plt.xticks(rotation=45)
```

Stability: The quality scores hover consistently around 2.5, with minor spikes and dips.

Outlier: A significant peak is visible around mid-October, indicating a notable increase in quality score for that period.

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=orig_data, x="ripeness_category", y="quality_score", hue="ripeness_category", dodge=False, palette="muted", inner="quartile", legend=False)
plt.title("Violin Plot")
plt.xlabel("Ripeness category")
plt.ylabel("Quality score")
plt.show()
```

This plot shows that that ovverripe bananas have highest quallity score, whereas green have the lowestv values

```python
plt.figure(figsize=(8, 6))
sns.histplot(orig_data["length_cm"], bins=50, kde=True, color="skyblue")
plt.title("Distribution of Banana's Length")
plt.xlabel("length_cm")
plt.ylabel("frequency")
plt.show()
```

This plot shows connection between frequency and length

The distribution is chaotic and no patterns have been identified 

The plot visualizes the distribution of quality_category across different regions, showing the count of products in each quality category: Premium, Good, and Unripe. The data excludes entries labeled as “Processing” due to the lack of relevant quality information.

```python
filtered_data = orig_data[orig_data["quality_category"] != "Processing"]
#I delete column 'Processing' because i we do not know any quallity information for this category

plt.figure(figsize=(12, 7))
sns.countplot(data=filtered_data, x="region", hue="quality_category", palette="viridis")
plt.title("Count of Quality Categories by Region")
plt.xlabel("Region")
plt.ylabel("Count")
plt.xticks(rotation=52)
plt.legend(title="Quality Category")
```

From the results, the majority of products across all regions fall under the Good category, with significantly fewer in Premium and Unripe categories. This suggests that the quality of products is consistently high across regions, but only a small portion achieves the Premium standard.

Hypothesis:

The high quality of bananas depends on sugar levels, ripeness, and soil nitrogen concentration, which collectively determine their taste and nutritional properties.

```python
filtered = orig_data.select_dtypes(include='number')
correlation = filtered.corr()

hypotesis = correlation['quality_score'].sort_values()
hypotesis = hypotesis.reset_index()
hypotesis.columns = ['Category','Correlation']
hypotesis = hypotesis[hypotesis['Category'] != 'quality_score']
hypotesis
```

```python
plt.figure(figsize=(14, 7))
sns.barplot(x='Category', y='Correlation', data=hypotesis, hue='Category', dodge=False, palette='coolwarm', legend=False)
plt.title('Correlation depending on a category')
plt.xticks(rotation=45, fontsize=12)
plt.tight_layout()
plt.show()
```

The bar plot shows the correlation of various features with a specific category. Among the variables, ripeness_index and sugar_content_brix exhibit the highest positive correlations, indicating their strong relevance to the category being analyzed.

This highlights the importance of ripeness and sugar content in distinguishing the categories.

```python
plt.figure(figsize=(10, 8))
scatter = sns.scatterplot(
    data=orig_data,
    x="ripeness_index",
    y="sugar_content_brix",
    hue="quality_category",
    size="firmness_kgf",
    sizes=(20, 200),
    alpha=0.9,
    palette="bright",
    marker="o"
)

plt.title("Ripeness Index vs Sugar Content Brix with Firmness as Size", fontsize=14)
plt.xlabel("Ripeness Index", fontsize=12)
plt.ylabel("Sugar Content (Brix)", fontsize=12)
plt.legend(title="Quality Category", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(alpha=0.3)
plt.tight_layout()

plt.show()
```

Sugar growth with a ripeness index: Riper bananas have more sugar.

Quality and ripeness: "Premium" bananas are more ripe and sweeter than "Processing".

Hardness and ripeness: More ripe bananas are less hard.

Category intersections: The quality categories partially coincide in terms of characteristics.
